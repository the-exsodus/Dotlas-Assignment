```python
from bs4 import BeautifulSoup as bs
import re
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
import json
```

```python
Header = ['restaurant_name', 'restaurant_logo', 'latitude', 'longitude', 'cuisine_tags', 'menu_items']

url = "https://www.talabat.com/uae/restaurant/611856/DOWNTOWN-BURJ-KHALIFA?&aid=1258"

options = Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
time.sleep(10)
page = driver.page_source
driver.quit()

webpage = bs(page, "html.parser")

print(webpage.prettify())
```

Restaurant Name

```python
restaurant_data = {}
restaurant_header = webpage.find("h1", attrs = {"class": "f-24 f-m-15 media-heading f-400 mb-0"})
restaurant_name = [text for text in restaurant_header.stripped_strings][0]
#print(restaurant_name)
restaurant_data[Header[0]] = restaurant_name
print(restaurant_data)
```

Restaurant Logo

```python
restaurant_img_block = webpage.find("div", attrs = {"class": "rest-logo-bg mr-4 mr-m-2 b-a"})
restaurant_img_url = restaurant_img_block.find("img")["src"]
restaurant_data[Header[1]] = restaurant_img_url
print(restaurant_data)
```

Latitude and Longitude

```python
coord_block = webpage.find("script", attrs = {"type" : "application/ld+json"}, string = re.compile("geo")).get_text()
#print(coord_block)
print("\n")
float_nums = re.findall(pattern= "[0-9]+[.][0-9]+", string  = coord_block)
lat = float(float_nums[0])
long = float(float_nums[1])
restaurant_data[Header[2]] = lat
restaurant_data[Header[3]] = long
print(restaurant_data)
```

Cuisine Tags

```python
cuisine_tags_str = webpage.find("p", attrs= {"class" : "light-text f-14 mb-0"}).get_text()
cuisine_tags_list = list(cuisine_tags_str.split(", "))
restaurant_data[Header[4]] = cuisine_tags_list
print(restaurant_data)
```

Menu Item List

```python
Menu_div = webpage.select("div.menu-category-containerstyles__MenuCategoryStyle-sc-l8657o-0.drAKkW")[0]
menu = Menu_div.find_all("div", attrs={"class" : "f-15"})
menu_items = []
for items in menu:
    menu_items.append(items.get_text())
#print(menu_items)
restaurant_data[Header[5]] = menu_items
print(restaurant_data)
```

Menu Dictionary

```python
menu_row = webpage.find_all("div", attrs = {"class" : "menu-itemstyles__MenuItemContainer-sc-wyx6cl-0"})
menu_item_list = []
```

```python
for index, row in enumerate(menu_row):
    item_info = {}
    item_name = menu_items[index]
    item_info["Item name"] = item_name
    item_desc = row.find("div", attrs = {"class" : "f-12 description"}).get_text()
    item_info["Item Description"] = item_desc
    item_price = row.find("div", attrs = {"class" : "mb-m-1"}).get_text()
    item_info["Item price"] = item_price
    item_image = row.find("img")["src"]
    item_info["Item image url"] = item_image
    menu_item_list.append(item_info)

print(menu_item_list)
```

```python
with open(restaurant_name + " details.txt", 'w') as write_file:
    write_file.write(json.dumps(restaurant_data))
```

```python
menu_data_df = pd.DataFrame(menu_item_list)
#print(menu_data_df)
menu_data_df.to_csv(restaurant_name + " menu details.csv")
```
